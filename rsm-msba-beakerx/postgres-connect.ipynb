```python
## psycopg2 does *not* return a pandas data.frame
import psycopg2
```

```python
#Define our connection string
conn_string = "host='postgres' dbname='postgres' user='postgres' password='postgres'"
conn = psycopg2.connect(conn_string)
```

```python
# conn.cursor will return a cursor object, you can use this cursor to perform queries
cursor = conn.cursor()
print ("Connected!\n")
```

```python
# execute our Query
cursor.execute("SELECT * FROM flights WHERE dep_time > 2350")
 
# retrieve the records from the database
records = cursor.fetchall()
#records
```
